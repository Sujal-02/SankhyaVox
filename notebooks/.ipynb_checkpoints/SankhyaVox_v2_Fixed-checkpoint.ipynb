# 🕉️ SankhyaVox v2 — Sanskrit Digit Recognition (0–99)
## End-to-End Pipeline: Real Data + Synthetic Data + Grammar-Constrained Viterbi

**Architecture:**
- 13 Sanskrit tokens → recognize any number 0–99
- Real recordings (your 5 speakers) + TTS synthetic data (gTTS + Edge-TTS)
- **Safe augmentation only** — pitch ±2 semitones, speed ±10%, isolated tokens ONLY
- HMM per token (Bakis, 3 states, 2 GMM) — right-sized for your dataset
- **Grammar-Constrained Viterbi Decoder** — grammar baked INTO decoding, not after

**Run cells top to bottom. Every cell prints verbose progress.**

## 📐 Step 2: Sanskrit Grammar (0–99)
> Defines every valid token sequence. Viterbi will use this to constrain decoding.


```python
# ═══════════════════════════════════════════════════════════════
#  STEP 2: Build Complete Grammar  (0 – 99)
# ═══════════════════════════════════════════════════════════════

UNITS = ["eka","dvi","tri","catur","pancha","shat","sapta","ashta","nava"]

def build_grammar():
    g = {}
    # 0
    g[0] = ["shunya"]
    # 1-9
    for i, u in enumerate(UNITS, 1):
        g[i] = [u]
    # 10
    g[10] = ["dasha"]
    # 11-19  →  dasha + unit
    for i, u in enumerate(UNITS, 1):
        g[10 + i] = ["dasha", u]
    # 20
    g[20] = ["vimsati"]
    # 21-29  →  vimsati + unit
    for i, u in enumerate(UNITS, 1):
        g[20 + i] = ["vimsati", u]
    # 30-99  →  multiplier + dasha [+ unit]
    tens_tok = {3:"tri", 4:"catur", 5:"pancha", 6:"shat",
                7:"sapta", 8:"ashta", 9:"nava"}
    for digit, mult in tens_tok.items():
        base = digit * 10
        g[base] = [mult, "dasha"]
        for i, u in enumerate(UNITS, 1):
            g[base + i] = [mult, "dasha", u]
    return g

GRAMMAR = build_grammar()

# Reverse: tuple(sequence) → integer
SEQ_TO_INT = {tuple(seq): val for val, seq in GRAMMAR.items()}

def tokens_to_number(token_list):
    return SEQ_TO_INT.get(tuple(token_list), -1)

# Build successor map:  token → set of tokens that can legally follow it
def build_successor_map():
    succ = {tok: set() for tok in VOCAB}
    succ["SIL"] = set(VOCAB)
    for val, seq in GRAMMAR.items():
        for i in range(len(seq) - 1):
            succ[seq[i]].add(seq[i+1])
        succ[seq[-1]].add("SIL")
    return succ

SUCCESSORS = build_successor_map()

# ── Verification ───────────────────────────────────────────────
assert tokens_to_number(["shunya"])              == 0
assert tokens_to_number(["pancha"])              == 5
assert tokens_to_number(["dasha", "pancha"])     == 15
assert tokens_to_number(["pancha", "dasha"])     == 50
assert tokens_to_number(["nava", "dasha", "nava"]) == 99
assert tokens_to_number(["vimsati", "eka"])      == 21

print(f"✅ Grammar built: {len(GRAMMAR)} valid sequences (0–99)")
print()
print("  Sample sequences:")
for n in [0, 5, 10, 15, 20, 21, 35, 50, 99]:
    print(f"   {n:3d}  →  {' + '.join(GRAMMAR[n])}")

print()
print("  Successor map (who can follow 'dasha'):", SUCCESSORS["dasha"])
print("  Successor map (who can follow 'pancha'):", SUCCESSORS["pancha"])
```

## 🔬 Step 3: Feature Extraction Functions
> Converts raw audio → 39-dimensional MFCC matrix (13 MFCC + 13 delta + 13 delta-delta).


```python
# ═══════════════════════════════════════════════════════════════
#  STEP 3: Feature Extraction  (MFCC 39-dim)
# ═══════════════════════════════════════════════════════════════

N_MFCC     = 13    # base MFCCs
N_FEATURES = 39    # 13 + 13 delta + 13 delta-delta
HOP_MS     = 10    # hop between frames in ms  →  100 frames/sec
WIN_MS     = 25    # analysis window in ms

def load_and_resample(path, target_sr=SAMPLE_RATE):
    """Load any audio file, resample to target_sr, convert to mono float32."""
    audio, sr = librosa.load(path, sr=target_sr, mono=True)
    return audio.astype(np.float32), target_sr

def preprocess_audio(audio, sr):
    """DC offset removal + peak normalisation."""
    audio = audio - np.mean(audio)              # remove DC offset
    peak  = np.max(np.abs(audio))
    if peak > 1e-6:
        audio = audio / peak * 0.95             # normalise to 95% peak
    return audio

def extract_mfcc(audio, sr):
    """
    Extract 39-dim MFCC matrix.
    Returns: np.ndarray  [T × 39]  (T = number of 10ms frames)
    """
    hop = int(sr * HOP_MS / 1000)    # 160 samples at 16kHz
    win = int(sr * WIN_MS / 1000)    # 400 samples at 16kHz

    # Pre-emphasis  →  boost high-freq consonants
    pre  = np.append(audio[0], audio[1:] - 0.97 * audio[:-1])

    mfcc = librosa.feature.mfcc(
        y=pre, sr=sr, n_mfcc=N_MFCC,
        hop_length=hop, n_fft=win
    )  # [13 × T]

    delta  = librosa.feature.delta(mfcc)           # velocity
    delta2 = librosa.feature.delta(mfcc, order=2)  # acceleration

    features = np.vstack([mfcc, delta, delta2])    # [39 × T]
    features = features.T                           # [T × 39]

    # CMVN — zero-mean, unit-variance per utterance
    mean = np.mean(features, axis=0, keepdims=True)
    std  = np.std(features,  axis=0, keepdims=True) + 1e-8
    features = (features - mean) / std

    return features.astype(np.float32)

def process_audio_file(wav_path):
    """Full pipeline: file path → 39-dim MFCC matrix."""
    audio, sr = load_and_resample(wav_path)
    audio     = preprocess_audio(audio, sr)
    mfcc      = extract_mfcc(audio, sr)
    return mfcc

# ── Quick self-test ────────────────────────────────────────────
_dummy = np.random.randn(SAMPLE_RATE).astype(np.float32)  # 1 sec of noise
_feats = extract_mfcc(_dummy, SAMPLE_RATE)
assert _feats.shape[1] == N_FEATURES, f"Expected 39 features, got {_feats.shape[1]}"
print(f"✅ Feature extractor OK")
print(f"   1 second of audio → MFCC matrix shape: {_feats.shape}")
print(f"   ({_feats.shape[0]} frames × {_feats.shape[1]} features  @ 100 fps)")
```

## 📂 Step 4: Load & Augment Real Recordings
> Loads your 5-speaker dataset. Safe augmentation: **pitch ±2, speed ±10% only.**
> No sequence splitting. No noise injection. Nothing that corrupts labels.


```python
# ═══════════════════════════════════════════════════════════════
#  STEP 4A: Load Real Recordings
# ═══════════════════════════════════════════════════════════════
# Expected folder structure:
#   data/real_recordings/
#       pancha/   pancha_s1_001.wav   pancha_s2_001.wav  ...
#       dasha/    dasha_s1_001.wav    ...
#       shunya/   shunya_s1_001.wav   ...
#       (one sub-folder named exactly like the token)
#
# If your files are named differently, the loader will still find them
# as long as they're inside a folder named after the token.

print("=" * 60)
print("  LOADING REAL RECORDINGS")
print("=" * 60)

real_audio = {tok: [] for tok in VOCAB}   # {token: [(audio, sr), ...]}

for tok in VOCAB:
    tok_dir = os.path.join(REAL_DATA_DIR, tok)
    if not os.path.isdir(tok_dir):
        print(f"  ⚠️  {tok:10s}  — folder not found: {tok_dir}")
        continue

    wavs = (glob.glob(os.path.join(tok_dir, "*.wav"))
          + glob.glob(os.path.join(tok_dir, "*.WAV"))
          + glob.glob(os.path.join(tok_dir, "*.mp3")))

    for wp in wavs:
        try:
            audio, sr = load_and_resample(wp)
            audio = preprocess_audio(audio, sr)
            real_audio[tok].append((audio, sr))
        except Exception as e:
            print(f"     ⚠️  Failed to load {os.path.basename(wp)}: {e}")

    count = len(real_audio[tok])
    bar   = "█" * count + "░" * max(0, 10 - count)
    print(f"  {tok:10s}  [{bar}]  {count:3d} recordings loaded")

total_real = sum(len(v) for v in real_audio.values())
print(f"\n✅ Total real recordings loaded: {total_real}")
if total_real == 0:
    print("\n❌  No real recordings found!")
    print(f"    Put your WAV files under: {REAL_DATA_DIR}/<token_name>/*.wav")
    print("    Example: data/real_recordings/pancha/pancha_s1_001.wav")
```

```python
# ═══════════════════════════════════════════════════════════════
#  STEP 4B: Safe Augmentation of Real Recordings
# ═══════════════════════════════════════════════════════════════
# ONLY isolated token augmentation.
# Pitch: -2, 0, +2 semitones  (3 variants — stays natural)
# Speed:  0.90, 1.0, 1.10     (3 variants — ±10%)
# Result: 3 × 3 = 9 variants per recording
# 50 real recordings × 9 aug = 450 samples per token

PITCH_STEPS = [-2, 0, 2]
SPEED_RATES = [0.90, 1.0, 1.10]

print("=" * 60)
print("  AUGMENTING REAL RECORDINGS  (pitch + speed only)")
print(f"  {len(PITCH_STEPS)} pitch variants × {len(SPEED_RATES)} speed variants = "
      f"{len(PITCH_STEPS)*len(SPEED_RATES)} augmentations per recording")
print("=" * 60)

augmented_audio = {tok: [] for tok in VOCAB}  # stores (audio, sr) tuples

for tok in VOCAB:
    recs = real_audio[tok]
    if not recs:
        print(f"  {tok:10s}  — skipped (no real data)")
        continue

    count = 0
    for audio, sr in recs:
        for pitch in PITCH_STEPS:
            for speed in SPEED_RATES:
                try:
                    y = audio.copy()
                    if speed != 1.0:
                        y = librosa.effects.time_stretch(y=y, rate=speed)
                    if pitch != 0:
                        y = librosa.effects.pitch_shift(y=y, sr=sr, n_steps=pitch)
                    # Trim silence edges (top_db=25 — gentle trim)
                    y_trimmed, _ = librosa.effects.trim(y, top_db=25)
                    if len(y_trimmed) > int(sr * 0.08):  # keep if > 80ms
                        y = y_trimmed
                    augmented_audio[tok].append((y.astype(np.float32), sr))
                    count += 1
                except Exception as e:
                    pass  # skip bad augmentation silently

    orig  = len(recs)
    total = len(augmented_audio[tok])
    bar   = "█" * min(total // 5, 20)
    print(f"  {tok:10s}  {orig:3d} real → {total:4d} augmented  [{bar}]")

total_aug = sum(len(v) for v in augmented_audio.values())
print(f"\n✅ Total after augmentation: {total_aug} samples across {len(VOCAB)} tokens")
```

## 🔊 Step 5: Synthetic Data Generation (gTTS + Edge-TTS)
> Generates TTS audio for each isolated token. **No multi-word sequences are ever generated.**
> Each TTS clip is also put through the safe augmentation (9 variants).


```python
# ═══════════════════════════════════════════════════════════════
#  STEP 5A: gTTS Synthesis  (5 accent variants per token)
# ═══════════════════════════════════════════════════════════════
from gtts import gTTS
import imageio_ffmpeg
ffmpeg_exe = imageio_ffmpeg.get_ffmpeg_exe()

GTTS_TLDS = ['co.in', 'com', 'co.uk', 'com.au', 'ca']  # 5 accent variants

synth_audio = {tok: [] for tok in VOCAB}   # will merge with augmented_audio later

print("=" * 60)
print("  gTTS SYNTHESIS")
print(f"  Generating {len(VOCAB)} tokens × {len(GTTS_TLDS)} accents")
print("=" * 60)

for tok in VOCAB:
    text = DEVANAGARI[tok]
    tok_count = 0
    print(f"  {tok:10s}  ({text})  ", end="", flush=True)

    for tld in GTTS_TLDS:
        tmp_mp3 = os.path.join(tempfile.gettempdir(), f"gtts_{tok}_{tld.replace('.','')}.mp3")
        tmp_wav = tmp_mp3.replace(".mp3", ".wav")
        try:
            gTTS(text=text, lang="hi", tld=tld).save(tmp_mp3)
            subprocess.run(
                [ffmpeg_exe, "-y", "-i", tmp_mp3,
                 "-ar", str(SAMPLE_RATE), "-ac", "1", tmp_wav],
                stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL, check=True
            )
            audio, sr = load_and_resample(tmp_wav)
            audio = preprocess_audio(audio, sr)
            synth_audio[tok].append((audio, sr))
            tok_count += 1
            print("g✓", end=" ", flush=True)
        except Exception as e:
            print("g✗", end=" ", flush=True)
        finally:
            for f in [tmp_mp3, tmp_wav]:
                if os.path.exists(f): os.remove(f)

    print(f"  →  {tok_count} clips")

total_gtts = sum(len(v) for v in synth_audio.values())
print(f"\n✅ gTTS generated: {total_gtts} clips")
```

```python
# ═══════════════════════════════════════════════════════════════
#  STEP 5B: Edge-TTS Synthesis  (2 Neural Hindi voices)
# ═══════════════════════════════════════════════════════════════
import asyncio, edge_tts

EDGE_VOICES = ["hi-IN-SwaraNeural", "hi-IN-MadhurNeural"]

print("=" * 60)
print("  Edge-TTS SYNTHESIS  (Neural Hindi voices)")
print(f"  Generating {len(VOCAB)} tokens × {len(EDGE_VOICES)} voices")
print("=" * 60)

async def edge_synthesize(text, voice, out_path):
    communicate = edge_tts.Communicate(text, voice)
    await communicate.save(out_path)

def run_edge_tts(text, voice, out_path):
    try:
        asyncio.get_event_loop().run_until_complete(
            edge_synthesize(text, voice, out_path)
        )
        return True
    except Exception:
        try:
            asyncio.run(edge_synthesize(text, voice, out_path))
            return True
        except Exception as e:
            return False

for tok in VOCAB:
    text = DEVANAGARI[tok]
    print(f"  {tok:10s}  ({text})  ", end="", flush=True)

    for voice in EDGE_VOICES:
        tmp_mp3 = os.path.join(tempfile.gettempdir(), f"edge_{tok}_{voice[:5]}.mp3")
        tmp_wav = tmp_mp3.replace(".mp3", ".wav")
        try:
            ok = run_edge_tts(text, voice, tmp_mp3)
            if not ok: raise RuntimeError("edge-tts failed")
            subprocess.run(
                [ffmpeg_exe, "-y", "-i", tmp_mp3,
                 "-ar", str(SAMPLE_RATE), "-ac", "1", tmp_wav],
                stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL, check=True
            )
            audio, sr = load_and_resample(tmp_wav)
            audio = preprocess_audio(audio, sr)
            synth_audio[tok].append((audio, sr))
            print("e✓", end=" ", flush=True)
        except Exception as e:
            print("e✗", end=" ", flush=True)
        finally:
            for f in [tmp_mp3, tmp_wav]:
                if os.path.exists(f): os.remove(f)

    print()

total_edge = sum(len(v) for v in synth_audio.values()) - total_gtts
print(f"\n✅ Edge-TTS added: {total_edge} clips")
print(f"   Total synthetic (gTTS + Edge): {sum(len(v) for v in synth_audio.values())} clips")
```

```python
# ═══════════════════════════════════════════════════════════════
#  STEP 5C: Augment Synthetic Clips (same safe matrix)
# ═══════════════════════════════════════════════════════════════
# 7 TTS clips × 9 augmentations = 63 synthetic variants per token
# Combined with ~450 real-augmented → ~513 total per token

print("=" * 60)
print("  AUGMENTING SYNTHETIC DATA")
print(f"  {len(PITCH_STEPS)} pitch × {len(SPEED_RATES)} speed = {len(PITCH_STEPS)*len(SPEED_RATES)} variants per clip")
print("=" * 60)

synth_augmented = {tok: [] for tok in VOCAB}

for tok in VOCAB:
    clips = synth_audio[tok]
    if not clips:
        print(f"  {tok:10s}  — no TTS clips available, skipping")
        continue

    for audio, sr in clips:
        for pitch in PITCH_STEPS:
            for speed in SPEED_RATES:
                try:
                    y = audio.copy()
                    if speed != 1.0:
                        y = librosa.effects.time_stretch(y=y, rate=speed)
                    if pitch != 0:
                        y = librosa.effects.pitch_shift(y=y, sr=sr, n_steps=pitch)
                    y_trimmed, _ = librosa.effects.trim(y, top_db=25)
                    if len(y_trimmed) > int(sr * 0.08):
                        y = y_trimmed
                    synth_augmented[tok].append((y.astype(np.float32), sr))
                except:
                    pass

    orig  = len(clips)
    total = len(synth_augmented[tok])
    print(f"  {tok:10s}  {orig:2d} TTS clips → {total:4d} augmented synthetic")

total_synth_aug = sum(len(v) for v in synth_augmented.values())
print(f"\n✅ Total augmented synthetic: {total_synth_aug}")
```

## 📊 Step 6: Extract Features from All Data (Real + Synthetic)
> Converts every audio clip to a 39-dim MFCC matrix and saves as `.npy`.


```python
# ═══════════════════════════════════════════════════════════════
#  STEP 6: Extract & Save MFCC Features for ALL data
# ═══════════════════════════════════════════════════════════════

print("=" * 60)
print("  FEATURE EXTRACTION")
print("=" * 60)

# token_features[tok] = list of np.ndarray [T × 39]
token_features   = {tok: [] for tok in VOCAB}
token_seq_lengths = {tok: [] for tok in VOCAB}

failed = 0

def extract_and_store(audio_list, tok, prefix, verbose_interval=50):
    """Extract MFCCs from a list of (audio, sr) and store in token_features."""
    count = 0
    for idx, (audio, sr) in enumerate(audio_list):
        try:
            mfcc = extract_mfcc(audio, sr)
            if mfcc.shape[0] < 5:   # too short, skip
                continue
            token_features[tok].append(mfcc)
            token_seq_lengths[tok].append(mfcc.shape[0])
            count += 1
            # Save to disk for reproducibility
            out_dir = os.path.join(FEATURE_DIR, tok)
            os.makedirs(out_dir, exist_ok=True)
            out_path = os.path.join(out_dir, f"{prefix}_{idx:04d}.npy")
            np.save(out_path, mfcc)
        except Exception as e:
            global failed
            failed += 1
    return count

for tok in VOCAB:
    n_real  = extract_and_store(augmented_audio[tok],    tok, "real_aug")
    n_synth = extract_and_store(synth_augmented[tok],    tok, "synth_aug")
    total   = n_real + n_synth
    avg_len = (np.mean(token_seq_lengths[tok]) if token_seq_lengths[tok] else 0)

    bar = "█" * min(total // 10, 25)
    print(f"  {tok:10s}  real:{n_real:4d}  synth:{n_synth:4d}  total:{total:5d}  "
          f"avg_frames:{avg_len:5.1f}  [{bar}]")

grand_total = sum(len(v) for v in token_features.values())
print(f"\n✅ Feature extraction complete")
print(f"   Grand total: {grand_total} MFCC matrices")
print(f"   Saved to:    {FEATURE_DIR}")
if failed:
    print(f"   ⚠️  {failed} clips failed extraction (too short or corrupt)")
```

## 🧠 Step 7: HMM Model Class
> Bakis (left-to-right) HMM per token. States and GMM components sized for your dataset.


```python
# ═══════════════════════════════════════════════════════════════
#  STEP 7: HMM Class  (FIXED: N_MIX=4 for problem tokens)
# ═══════════════════════════════════════════════════════════════
from hmmlearn.hmm import GMMHMM

# State count per token — sized to phonetic complexity
HMM_STATES = {
    "shunya":  5,
    "eka":     4,   # bumped from 3 — was confused with tri
    "dvi":     4,   # bumped from 3 — was confused with tri
    "tri":     3,
    "catur":   5,
    "pancha":  5,
    "shat":    4,   # bumped from 3 — was confused with shata
    "sapta":   5,
    "ashta":   5,
    "nava":    4,
    "dasha":   4,
    "vimsati": 6,
    "shata":   4,
}

# Per-token GMM mixtures — more for confused tokens
HMM_MIX = {
    "shunya":  2,
    "eka":     4,   # confused with tri/pancha — more mixtures
    "dvi":     4,   # confused with tri — more mixtures
    "tri":     3,
    "catur":   2,
    "pancha":  4,   # confused with eka — more mixtures
    "shat":    4,   # confused with shata — more mixtures
    "sapta":   2,
    "ashta":   2,
    "nava":    2,
    "dasha":   2,
    "vimsati": 2,
    "shata":   4,   # confused with shat — more mixtures
}

N_ITER = 150  # more EM iterations (was 100)

class SankhyaHMM:
    """One GMM-HMM per Sanskrit token. Bakis (left-to-right) topology."""

    def __init__(self, vocab, states_map, model_dir, mix_map=None, n_iter=N_ITER):
        self.vocab     = vocab
        self.states    = states_map
        self.mix_map   = mix_map or {t: 2 for t in vocab}
        self.model_dir = model_dir
        self.n_iter    = n_iter
        self.models    = {}

    def _make_bakis_transmat(self, n_states):
        """Left-to-right: stay in state or advance. No skipping, no backward."""
        A = np.zeros((n_states, n_states))
        for i in range(n_states - 1):
            A[i, i]   = 0.6
            A[i, i+1] = 0.4
        A[-1, -1] = 1.0
        return A

    def fit(self, token, sequences, lengths):
        n_states = self.states[token]
        n_mix    = self.mix_map[token]
        model = GMMHMM(
            n_components    = n_states,
            n_mix           = n_mix,
            n_iter          = self.n_iter,
            covariance_type = "diag",
            verbose         = False,
        )
        model.startprob_prior         = np.zeros(n_states) + 1e-10
        model.startprob_prior[0]      = 1.0
        model.transmat_prior          = self._make_bakis_transmat(n_states) + 1e-2
        model.fit(sequences, lengths)
        self.models[token] = model

    def score(self, token, mfcc_matrix):
        """Per-FRAME log-likelihood (raw LL divided by sequence length)."""
        if token not in self.models:
            return -1e9
        try:
            ll = self.models[token].score(mfcc_matrix)
            return ll / max(len(mfcc_matrix), 1)   # ← KEY: per-frame normalise
        except:
            return -1e9

    def save(self):
        os.makedirs(self.model_dir, exist_ok=True)
        for tok, model in self.models.items():
            joblib.dump(model, os.path.join(self.model_dir, f"hmm_{tok}.pkl"))
        print(f"  ✅ Saved {len(self.models)} HMM models to {self.model_dir}")

    def load(self):
        self.models = {}
        for tok in self.vocab:
            path = os.path.join(self.model_dir, f"hmm_{tok}.pkl")
            if os.path.exists(path):
                self.models[tok] = joblib.load(path)
        print(f"  ✅ Loaded {len(self.models)} HMM models from {self.model_dir}")
        return self


print("✅ SankhyaHMM class defined  (per-frame score normalisation)")
print()
print(f"  {'Token':10s}  {'States':>6}  {'GMM Mix':>7}")
print(f"  {'-'*30}")
for t in VOCAB:
    flag = " ← increased" if HMM_MIX[t] > 2 else ""
    print(f"  {t:10s}  {HMM_STATES[t]:>6}  {HMM_MIX[t]:>7}{flag}")
print(f"\n  EM iterations: {N_ITER}")
```

## 🏋️ Step 8: Train HMMs


```python
# ═══════════════════════════════════════════════════════════════
#  STEP 8: Train One HMM Per Token
# ═══════════════════════════════════════════════════════════════

hmm_bank = SankhyaHMM(VOCAB, HMM_STATES, MODEL_DIR, mix_map=HMM_MIX, n_iter=N_ITER)

print("=" * 60)
print("  HMM TRAINING")
print(f"  {len(VOCAB)} tokens  ×  Bakis topology  ×  per-token GMM mixtures")
print("=" * 60)

train_log = {}
t0_global = time.time()

for tok in VOCAB:
    seqs    = token_features[tok]
    lengths = token_seq_lengths[tok]

    if not seqs:
        print(f"  {tok:10s}  ❌  NO DATA — skipping")
        continue

    n_samples    = len(seqs)
    total_frames = sum(lengths)
    X_concat     = np.concatenate(seqs, axis=0)

    print(f"  {tok:10s}  n={n_samples:4d}  frames={total_frames:6d}  "
          f"states={HMM_STATES[tok]}  mix={HMM_MIX[tok]}  ", end="", flush=True)

    t0 = time.time()
    try:
        hmm_bank.fit(tok, X_concat, lengths)
        elapsed = time.time() - t0

        # Score a few training samples (per-frame, so comparable across tokens)
        sample_scores = [hmm_bank.score(tok, s) for s in seqs[:min(5, len(seqs))]]
        avg_score     = np.mean(sample_scores)

        train_log[tok] = {
            "n_samples":     n_samples,
            "total_frames":  int(total_frames),
            "avg_perframe_loglik": round(float(avg_score), 4),
            "train_sec":     round(elapsed, 2),
        }
        print(f"✅  {elapsed:.1f}s  per-frame-LL={avg_score:.3f}")

    except Exception as e:
        print(f"❌  ERROR: {e}")
        train_log[tok] = {"error": str(e)}

total_time = time.time() - t0_global
print(f"\n✅ Training complete in {total_time:.1f}s  ({total_time/60:.1f} min)")
hmm_bank.save()

with open(os.path.join(MODEL_DIR, "train_log.json"), "w") as f:
    json.dump(train_log, f, indent=2)
print(f"   Log saved: {MODEL_DIR}/train_log.json")

# ── Quick per-token score check ─────────────────────────────
print()
print("  Per-frame log-likelihood on training samples:")
print(f"  {'Token':10s}  {'per-frame LL':>13}  Note")
for tok, info in train_log.items():
    if "error" in info: continue
    ll = info["avg_perframe_loglik"]
    note = "⚠️  unusually low" if ll < -60 else ("✅" if ll > -45 else "ok")
    print(f"  {tok:10s}  {ll:>13.3f}  {note}")
print()
print("  Tip: scores should be in the -30 to -55 range.")
print("  If a token is much lower, it may have bad training data.")
```

## 🎯 Step 9: Grammar-Constrained Viterbi Decoder (v2 — Fixed)
> **Fixes:** SIL state removed (was dominating scores), per-frame HMM normalisation, energy-threshold silence gate before Viterbi.


```python
# ═══════════════════════════════════════════════════════════════
#  STEP 9: Grammar-Constrained Viterbi Decoder  (FIXED)
#
#  Key fixes vs v1:
#   1. SIL state REMOVED — it was dominating scores at wrong scale
#   2. HMM.score() already returns PER-FRAME LL (fixed in HMM class)
#   3. Energy threshold pre-filters silent windows before Viterbi
#   4. Cleaner transition matrix (token→token only, no SIL states)
# ═══════════════════════════════════════════════════════════════

class GrammarConstrainedDecoder:
    """
    Decodes continuous Sanskrit digit speech → integer (0–99).

    Pipeline:
      1. Slide 250ms window every 50ms over MFCC matrix
      2. For each window: compute per-frame LL for all 13 HMMs
         → Score matrix [N_windows × 13]
      3. Mark windows below energy threshold as SILENT (skip them)
      4. Grammar-constrained Viterbi over non-silent windows
         (impossible token transitions pruned during search)
      5. Backtrack → token sequence → grammar lookup → integer

    Why no SIL state:
      HMM log-likelihoods are per-frame values (~-35 to -55).
      Any energy-based SIL score uses a completely different scale
      and will dominate or be dominated unpredictably.
      We handle silence via energy threshold BEFORE Viterbi instead.
    """

    WINDOW_FRAMES  = 25    # 250ms at 100fps
    HOP_FRAMES     = 5     # 50ms hop
    SILENCE_DB     = -35   # windows with energy below this (dBFS) are skipped
    MIN_TOKEN_WINS = 3     # a token must appear in ≥ this many windows to count

    def __init__(self, hmm_bank, grammar, successors, vocab):
        self.hmm     = hmm_bank
        self.grammar = grammar
        self.succ    = successors
        self.vocab   = vocab
        self.n       = len(vocab)

        self.tok2idx = {t: i for i, t in enumerate(vocab)}
        self.idx2tok = {i: t for t, i in self.tok2idx.items()}

        # Boolean transition matrix [n × n]
        # allowed[i,j] = True if vocab[j] can legally follow vocab[i]
        self.allowed = np.zeros((self.n, self.n), dtype=bool)
        # Any token can START (first window = free choice)
        self._start_allowed = np.ones(self.n, dtype=bool)

        for tok, succs in self.succ.items():
            i = self.tok2idx.get(tok)
            if i is None:
                continue
            for s in succs:
                j = self.tok2idx.get(s)
                if j is not None:
                    self.allowed[i, j] = True
        # Self-loops allowed (stay in same token across consecutive windows)
        np.fill_diagonal(self.allowed, True)

    # ── 1. Score Matrix ────────────────────────────────────────
    def compute_scores(self, mfcc):
        """
        Slide window, score each against all HMMs.

        Returns:
            scores:   [N × n_vocab]  per-frame log-likelihoods
            is_sil:   [N]  bool array — True if window is silence
        """
        T = mfcc.shape[0]
        W = self.WINDOW_FRAMES
        H = self.HOP_FRAMES

        positions = list(range(0, max(1, T - W + 1), H))

        N      = len(positions)
        scores = np.full((N, self.n), -1e9, dtype=np.float64)
        is_sil = np.zeros(N, dtype=bool)

        for wi, start in enumerate(positions):
            window = mfcc[start : start + W]
            if window.shape[0] < W:
                window = np.pad(window, ((0, W - window.shape[0]), (0, 0)), mode="edge")

            # Energy check — skip silence windows
            rms_db = 20 * np.log10(np.sqrt(np.mean(window ** 2)) + 1e-10)
            if rms_db < self.SILENCE_DB:
                is_sil[wi] = True
                continue

            for ti, tok in enumerate(self.vocab):
                scores[wi, ti] = self.hmm.score(tok, window)  # already per-frame

        return scores, is_sil

    # ── 2. Viterbi ─────────────────────────────────────────────
    def viterbi(self, scores, is_sil):
        """
        Grammar-constrained Viterbi. Silent windows are skipped.

        Returns:
            collapsed: list of token strings (consecutive dupes merged)
            path_score: float
        """
        # Filter out silent windows
        active_idx = np.where(~is_sil)[0]

        if len(active_idx) == 0:
            return [], 0.0

        active_scores = scores[active_idx]   # [M × n]
        M = len(active_idx)

        vit  = np.full((M, self.n), -np.inf)
        back = np.full((M, self.n), -1, dtype=np.int32)

        # Initialise — any token can start
        vit[0] = active_scores[0]

        # DP
        for t in range(1, M):
            for s in range(self.n):
                emit = active_scores[t, s]
                if emit < -1e8:
                    continue
                # Best predecessor that can transition to s
                prev_scores = vit[t-1].copy()
                prev_scores[~self.allowed[:, s]] = -np.inf
                best_prev = int(np.argmax(prev_scores))
                if prev_scores[best_prev] > -np.inf:
                    vit[t, s]  = prev_scores[best_prev] + emit
                    back[t, s] = best_prev

        # Best end state
        end_s      = int(np.argmax(vit[M-1]))
        path_score = float(vit[M-1, end_s])

        if path_score == -np.inf:
            return [], -np.inf

        # Backtrack
        path = [end_s]
        s = end_s
        for t in range(M-1, 0, -1):
            s = int(back[t, s])
            if s < 0:
                break
            path.append(s)
        path.reverse()

        # Collapse consecutive identical tokens
        collapsed = []
        for s in path:
            tok = self.idx2tok[s]
            if not collapsed or collapsed[-1] != tok:
                collapsed.append(tok)

        return collapsed, path_score

    # ── 3. Grammar lookup ──────────────────────────────────────
    def sequence_to_int(self, tokens):
        """Exact match first, then try contiguous sub-sequences."""
        r = tokens_to_number(tokens)
        if r >= 0:
            return r
        for length in range(1, min(4, len(tokens) + 1)):
            for start in range(len(tokens) - length + 1):
                r = tokens_to_number(tokens[start : start + length])
                if r >= 0:
                    return r
        return -1

    # ── 4. Full decode ─────────────────────────────────────────
    def decode(self, mfcc, verbose=False):
        """
        MFCC matrix [T × 39] → (integer, token_list, debug_dict)
        Returns integer=-1 if recognition fails.
        """
        if verbose:
            print(f"  Input MFCC : {mfcc.shape}  ({mfcc.shape[0]/100:.2f}s)")

        scores, is_sil = self.compute_scores(mfcc)
        n_active = int((~is_sil).sum())

        if verbose:
            print(f"  Score matrix: {scores.shape}")
            print(f"  Active (non-silent) windows: {n_active} / {len(is_sil)}")
            if n_active > 0:
                best_per_win = [
                    self.idx2tok[int(np.argmax(scores[i]))]
                    if not is_sil[i] else "SIL"
                    for i in range(len(is_sil))
                ]
                print(f"  Best token per window: {best_per_win}")

        token_seq, score = self.viterbi(scores, is_sil)

        if verbose:
            print(f"  Viterbi path : {token_seq}  (score={score:.3f})")

        integer_result = self.sequence_to_int(token_seq)

        if verbose:
            print(f"  Grammar      : {token_seq} → {integer_result}")

        debug = {
            "score_matrix_shape": list(scores.shape),
            "active_windows":     n_active,
            "raw_token_sequence": token_seq,
            "viterbi_score":      round(score, 4),
            "recognized_integer": integer_result,
        }
        return integer_result, token_seq, debug


# ── Instantiate ────────────────────────────────────────────────
decoder = GrammarConstrainedDecoder(
    hmm_bank   = hmm_bank,
    grammar    = GRAMMAR,
    successors = SUCCESSORS,
    vocab      = VOCAB,
)

print("✅ GrammarConstrainedDecoder  (v2 — SIL removed, per-frame scores)")
print(f"   Vocab size    : {decoder.n} tokens")
print(f"   Window        : {decoder.WINDOW_FRAMES} frames (250ms)")
print(f"   Hop           : {decoder.HOP_FRAMES} frames  (50ms)")
print(f"   Silence gate  : {decoder.SILENCE_DB} dBFS")
print(f"   Transitions   : {decoder.allowed.sum()} allowed "
      f"out of {decoder.allowed.size} possible")
print()
print("  Transition map (sample):")
for tok in ["dasha", "pancha", "vimsati", "shunya"]:
    succs = [VOCAB[j] for j in range(decoder.n) if decoder.allowed[decoder.tok2idx[tok], j]
             and VOCAB[j] != tok]
    print(f"    {tok:10s} → {succs}")
```

## 🧪 Step 10: Evaluation
> Tests isolated token recognition + compound number recognition.


```python
# ═══════════════════════════════════════════════════════════════
#  STEP 10A: Isolated Token Recognition Accuracy
# ═══════════════════════════════════════════════════════════════
# Tests: given a known single-token MFCC, does the HMM rank it
# #1 among all 13 tokens?

print("=" * 60)
print("  ISOLATED TOKEN RECOGNITION  (hold-out evaluation)")
print("=" * 60)
print(f"  {'Token':10s}  {'Correct':>8}  {'Total':>7}  {'Acc%':>6}  Confusion")
print("-" * 60)

per_token_results = {}
confusion = np.zeros((len(VOCAB), len(VOCAB)), dtype=int)

# Use last 20% of samples as hold-out (indices from the end)
for tok in VOCAB:
    seqs = token_features[tok]
    if not seqs:
        print(f"  {tok:10s}  — no data")
        per_token_results[tok] = {"acc": 0.0, "n": 0}
        continue

    n_total   = len(seqs)
    n_holdout = max(1, int(n_total * 0.20))
    holdout   = seqs[-n_holdout:]   # last 20%

    true_idx  = TOKEN_TO_IDX[tok]
    correct   = 0
    wrong_preds = []

    for mfcc_seq in holdout:
        # Score against all tokens
        all_scores = {t: hmm_bank.score(t, mfcc_seq) for t in VOCAB}
        best_tok   = max(all_scores, key=all_scores.get)
        pred_idx   = TOKEN_TO_IDX[best_tok]
        confusion[true_idx, pred_idx] += 1

        if best_tok == tok:
            correct += 1
        else:
            wrong_preds.append(best_tok)

    acc = correct / len(holdout) * 100
    per_token_results[tok] = {"acc": acc, "n": len(holdout)}

    wrong_str = ", ".join(f"{w}({wrong_preds.count(w)})"
                          for w in sorted(set(wrong_preds))) or "none"
    status = "✅" if acc >= 80 else ("⚠️ " if acc >= 60 else "❌")
    print(f"  {tok:10s}  {correct:>6}/{len(holdout):<5}  {acc:>5.1f}%  {status}  "
          f"{'confused with: ' + wrong_str if wrong_preds else ''}")

mean_acc = np.mean([v["acc"] for v in per_token_results.values() if v["n"] > 0])
print("-" * 60)
print(f"  Mean accuracy: {mean_acc:.1f}%")
if mean_acc >= 85:
    print("  ✅ Excellent! Ready for continuous speech testing.")
elif mean_acc >= 70:
    print("  ⚠️  Good. Collect more real recordings for tokens below 80%.")
else:
    print("  ❌  Low. Check that REAL_DATA_DIR/token/ folders exist and have WAVs.")
```

```python
# ═══════════════════════════════════════════════════════════════
#  STEP 10B: Confusion Matrix Heatmap
# ═══════════════════════════════════════════════════════════════

fig, ax = plt.subplots(figsize=(11, 9))
im = ax.imshow(confusion, cmap="Blues")
ax.set_xticks(range(len(VOCAB))); ax.set_xticklabels(VOCAB, rotation=45, ha="right")
ax.set_yticks(range(len(VOCAB))); ax.set_yticklabels(VOCAB)
ax.set_xlabel("Predicted"); ax.set_ylabel("True")
ax.set_title("Confusion Matrix — Isolated Token Recognition", fontsize=13)
plt.colorbar(im, ax=ax)
for i in range(len(VOCAB)):
    for j in range(len(VOCAB)):
        if confusion[i, j] > 0:
            ax.text(j, i, str(confusion[i, j]), ha="center", va="center",
                    color="white" if confusion[i,j] > confusion.max()*0.5 else "black",
                    fontsize=8)
plt.tight_layout()
plt.savefig(os.path.join(PROJECT_ROOT, "confusion_matrix.png"), dpi=120)
plt.show()
print("Confusion matrix saved.")
```

```python
# ═══════════════════════════════════════════════════════════════
#  STEP 10C: Compound Number Recognition  (FIXED)
#
#  OLD approach (broken): synthesise full phrase "दश पञ्च" via gTTS
#    → Connected Hindi TTS sounds nothing like two isolated Sanskrit
#      words — the HMMs were trained on isolated tokens, so they
#      cannot decode connected TTS speech.
#
#  NEW approach: concatenate real isolated token WAVs with 80ms silence
#    → Exactly mimics a user saying "pancha ... dasha" naturally.
#    → If real_audio has samples for both tokens, use them.
#    → Falls back to gTTS isolated clips if real data missing.
# ═══════════════════════════════════════════════════════════════

print("=" * 60)
print("  COMPOUND NUMBER RECOGNITION")
print("  (concatenated isolated WAVs + 80ms silence gap)")
print("=" * 60)

SILENCE_GAP_MS = 80   # ms of silence between tokens in test

def get_one_clip(tok):
    """Return one (audio, sr) clip for tok — real preferred, then synth."""
    pool = real_audio.get(tok, []) or synth_audio.get(tok, [])
    if not pool:
        return None, None
    audio, sr = pool[0]
    return audio.copy(), sr

def build_compound_audio(token_sequence, gap_ms=SILENCE_GAP_MS):
    """
    Concatenate audio clips for each token in sequence,
    separated by gap_ms of silence.
    Returns (audio_array, sample_rate) or (None, None) if any token missing.
    """
    parts = []
    sr_ref = SAMPLE_RATE
    gap = np.zeros(int(sr_ref * gap_ms / 1000), dtype=np.float32)

    for i, tok in enumerate(token_sequence):
        audio, sr = get_one_clip(tok)
        if audio is None:
            return None, None
        if sr != sr_ref:
            audio = librosa.resample(audio, orig_sr=sr, target_sr=sr_ref)
        parts.append(audio)
        if i < len(token_sequence) - 1:
            parts.append(gap)   # silence between words

    return np.concatenate(parts).astype(np.float32), sr_ref


TEST_NUMBERS = [0, 5, 10, 15, 21, 35, 42, 50, 63, 71, 88, 99]
compound_results = []

print(f"  {'Target':>7}  {'Token Sequence':30s}  {'Pred':>6}  {'Status':6}  Detail")
print(f"  {'-'*75}")

for target in TEST_NUMBERS:
    seq = GRAMMAR[target]
    compound, sr = build_compound_audio(seq)

    if compound is None:
        print(f"  {target:>7}  {str(seq):30s}  {'N/A':>6}  ⚠️      missing clip for a token")
        continue

    compound = preprocess_audio(compound, sr)
    mfcc     = extract_mfcc(compound, sr)

    pred_int, pred_toks, debug = decoder.decode(mfcc, verbose=False)
    ok = (pred_int == target)
    compound_results.append(ok)

    status = "✅" if ok else "❌"
    detail = f"decoded as {pred_toks}" if not ok else ""
    print(f"  {target:>7}  {str(seq):30s}  {pred_int:>6}  {status}      {detail}")

correct = sum(compound_results)
total   = len(compound_results)
pct     = correct / max(total, 1) * 100
print(f"\n  Result: {correct}/{total} = {pct:.1f}%")

if pct == 0:
    print()
    print("  ── DEBUG: Verbose decode on first failed number ──")
    for target in TEST_NUMBERS:
        seq = GRAMMAR[target]
        compound, sr = build_compound_audio(seq)
        if compound is None: continue
        compound = preprocess_audio(compound, sr)
        mfcc     = extract_mfcc(compound, sr)
        print(f"\n  Decoding {target} = {seq}")
        decoder.decode(mfcc, verbose=True)
        break
```
