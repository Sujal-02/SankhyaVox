## 🗺️ Step 2: Text Maps & Augmentation Engine

```python
# Devanagari text for single-token TTS
DEVANAGARI_MAP = {
    "shunya": "शून्य", "eka": "एक", "dvi": "द्वि", "tri": "त्रि",
    "catur": "चतुर", "pancha": "पञ्च", "shat": "षट", "sapta": "सप्त",
    "ashta": "अष्ट", "nava": "नव", "dasha": "दश",
    "vimsati": "विंशति", "shata": "शत",
}

# Multi-word sequences for double-digit training (CRITICAL!)
SEQUENCE_MAP = {
    "dasha_eka": "दश एक", "dasha_dvi": "दश द्वि", "dasha_tri": "दश त्रि",
    "dasha_catur": "दश चतुर", "dasha_pancha": "दश पञ्च", "dasha_shat": "दश षट",
    "dasha_sapta": "दश सप्त", "dasha_ashta": "दश अष्ट", "dasha_nava": "दश नव",
    "vimsati_eka": "विंशति एक", "vimsati_pancha": "विंशति पञ्च",
    "vimsati_sapta": "विंशति सप्त", "vimsati_nava": "विंशति नव",
    "tri_dasha": "त्रि दश", "tri_dasha_pancha": "त्रि दश पञ्च",
    "catur_dasha": "चतुर दश", "catur_dasha_dvi": "चतुर दश द्वि",
    "pancha_dasha": "पञ्च दश", "pancha_dasha_sapta": "पञ्च दश सप्त",
    "shat_dasha": "षट दश", "shat_dasha_tri": "षट दश त्रि",
    "sapta_dasha": "सप्त दश", "sapta_dasha_dvi": "सप्त दश द्वि",
    "ashta_dasha": "अष्ट दश", "ashta_dasha_pancha": "अष्ट दश पञ्च",
    "nava_dasha": "नव दश", "nava_dasha_nava": "नव दश नव",
}

# Augmentation matrix
PITCH_STEPS  = [-4, -2, 0, 2, 4]     # 5 pitch variants
SPEED_RATES  = [0.85, 1.0, 1.15]      # 3 speed variants
NOISE_LEVELS = [0.0, 0.003]            # clean + noise

generated_count = 0

def augment_and_save(audio, sr, token, var_name):
    """Apply augmentation matrix to single-token audio."""
    global generated_count
    trimmed, _ = librosa.effects.trim(audio, top_db=30)
    if len(trimmed) > int(sr * 0.1): audio = trimmed
    for p in PITCH_STEPS:
        for s in SPEED_RATES:
            for n in NOISE_LEVELS:
                y = audio.copy()
                if s != 1.0: y = librosa.effects.time_stretch(y=y, rate=s)
                if p != 0: y = librosa.effects.pitch_shift(y=y, sr=sr, n_steps=p)
                if n > 0: y = y + np.random.normal(0, n, len(y)).astype(np.float32)
                out = os.path.join(RAW_SYNTH_DIR, f"synth_{var_name}_p{p}_s{int(s*100)}_n{int(n*10000)}_{token}.wav")
                sf.write(out, y.astype(np.float32), sr)
                try:
                    process_file(out, os.path.join(str(FEATURE_DIR), f"synth_{var_name}"))
                    generated_count += 1
                except: pass

def augment_sequence(audio, sr, tokens, var_name):
    """Split multi-word audio into individual token segments + augment."""
    global generated_count
    trimmed, _ = librosa.effects.trim(audio, top_db=30)
    if len(trimmed) > int(sr * 0.1): audio = trimmed
    n_tok = len(tokens)
    for p in PITCH_STEPS:
        for s in SPEED_RATES:
            for n in NOISE_LEVELS:
                y = audio.copy()
                if s != 1.0: y = librosa.effects.time_stretch(y=y, rate=s)
                if p != 0: y = librosa.effects.pitch_shift(y=y, sr=sr, n_steps=p)
                if n > 0: y = y + np.random.normal(0, n, len(y)).astype(np.float32)
                seg = len(y) // n_tok
                for i, tok in enumerate(tokens):
                    chunk = y[i*seg : (i+1)*seg if i < n_tok-1 else len(y)]
                    out = os.path.join(RAW_SYNTH_DIR, f"seq_{var_name}_p{p}_s{int(s*100)}_n{int(n*10000)}_{tok}.wav")
                    sf.write(out, chunk.astype(np.float32), sr)
                    try:
                        process_file(out, os.path.join(str(FEATURE_DIR), f"seq_{var_name}"))
                        generated_count += 1
                    except: pass

augs = len(PITCH_STEPS) * len(SPEED_RATES) * len(NOISE_LEVELS)
print(f"✅ Augmentation engine: {augs} variants per voice")
print(f"✅ {len(DEVANAGARI_MAP)} single tokens + {len(SEQUENCE_MAP)} multi-word sequences")
```

## 🔊 Step 3: Generate — Google TTS (5 accents)

```python
from gtts import gTTS

GTTS_TLDS = ['co.in', 'com', 'co.uk', 'com.au', 'ca']

print("=== gTTS: Single Tokens ===")
for token in VOCAB:
    if token not in DEVANAGARI_MAP: continue
    for tld in GTTS_TLDS:
        print(f"  {token} ({tld})", end=".. ", flush=True)
        with tempfile.NamedTemporaryFile(suffix='.mp3', delete=False) as f: tmp = f.name
        try:
            gTTS(text=DEVANAGARI_MAP[token], lang='hi', tld=tld).save(tmp)
            wav = tmp.replace('.mp3','.wav')
            subprocess.run([ffmpeg_exe,'-y','-i',tmp,'-ar',str(SAMPLE_RATE),'-ac','1',wav],
                          stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL, check=True)
            a, sr = sf.read(wav); augment_and_save(a, sr, token, f"gtts_{tld.replace('.','')}"); os.remove(wav)
            print("✓", end=" ")
        except Exception as e: print(f"✗", end=" ")
        finally:
            if os.path.exists(tmp): os.remove(tmp)
    print()
print(f"\n📊 Running total: {generated_count} features")
```

---
## 🔊 Step 4: Generate — Microsoft Edge Neural TTS (2 voices)

```python
EDGE_VOICES = ['hi-IN-SwaraNeural', 'hi-IN-MadhurNeural']

print("=== Edge-TTS: Single Tokens ===")
for token in VOCAB:
    if token not in DEVANAGARI_MAP: continue
    for voice in EDGE_VOICES:
        print(f"  {token} ({voice.split('-')[-1]})", end=".. ", flush=True)
        with tempfile.NamedTemporaryFile(suffix='.mp3', delete=False) as f: tmp = f.name
        try:
            subprocess.run(['edge-tts','--voice',voice,'--text',DEVANAGARI_MAP[token],'--write-media',tmp],
                          stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL, check=True)
            wav = tmp.replace('.mp3','.wav')
            subprocess.run([ffmpeg_exe,'-y','-i',tmp,'-ar',str(SAMPLE_RATE),'-ac','1',wav],
                          stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL, check=True)
            a, sr = sf.read(wav); augment_and_save(a, sr, token, voice.replace('-','')); os.remove(wav)
            print("✓", end=" ")
        except: print("✗", end=" ")
        finally:
            if os.path.exists(tmp): os.remove(tmp)
    print()
print(f"\n📊 Running total: {generated_count} features")
```

---
## 🔊 Step 5: Generate — pyttsx3 Offline (Windows SAPI5)

```python
print("=== pyttsx3: Offline SAPI5 ===")
try:
    import pyttsx3
    engine = pyttsx3.init()
    engine.setProperty('rate', 150)
    for token in VOCAB:
        if token not in DEVANAGARI_MAP: continue
        print(f"  {token}", end=".. ", flush=True)
        tmp = os.path.join(tempfile.gettempdir(), f"pyttsx3_{token}.wav")
        engine.save_to_file(DEVANAGARI_MAP[token], tmp)
        engine.runAndWait()
        if os.path.exists(tmp) and os.path.getsize(tmp) > 1000:
            a, sr = sf.read(tmp)
            if sr != SAMPLE_RATE: a = librosa.resample(a, orig_sr=sr, target_sr=SAMPLE_RATE)
            augment_and_save(a, SAMPLE_RATE, token, 'pyttsx3')
            print("✓", end=" ")
        else: print("✗", end=" ")
        if os.path.exists(tmp): os.remove(tmp)
    engine.stop()
    print()
except Exception as e: print(f"⚠️ pyttsx3 skipped: {e}")
print(f"\n📊 Running total: {generated_count} features")
```

## 🔗 Step 7: Generate Multi-Word Sequences (CRITICAL for double digits!)
This synthesizes full phrases like "दश पञ्च" (15) and splits them into word-level segments.
Without this, the HMM cannot accurately decode two-digit numbers.

```python
print("=== Multi-Word Sequences ===")
for seq_key, seq_text in SEQUENCE_MAP.items():
    tokens = seq_key.split("_")
    print(f"  {seq_key}", end=": ")
    # gTTS
    for tld in ['co.in', 'com']:
        with tempfile.NamedTemporaryFile(suffix='.mp3', delete=False) as f: tmp = f.name
        try:
            gTTS(text=seq_text, lang='hi', tld=tld).save(tmp)
            wav = tmp.replace('.mp3','.wav')
            subprocess.run([ffmpeg_exe,'-y','-i',tmp,'-ar',str(SAMPLE_RATE),'-ac','1',wav],
                          stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL, check=True)
            a, sr = sf.read(wav); augment_sequence(a, sr, tokens, f"gtts_{tld.replace('.','')}_seq"); os.remove(wav)
            print(f"g✓", end=" ")
        except: print("g✗", end=" ")
        finally:
            if os.path.exists(tmp): os.remove(tmp)
    # Edge-TTS
    for voice in EDGE_VOICES:
        with tempfile.NamedTemporaryFile(suffix='.mp3', delete=False) as f: tmp = f.name
        try:
            subprocess.run(['edge-tts','--voice',voice,'--text',seq_text,'--write-media',tmp],
                          stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL, check=True)
            wav = tmp.replace('.mp3','.wav')
            subprocess.run([ffmpeg_exe,'-y','-i',tmp,'-ar',str(SAMPLE_RATE),'-ac','1',wav],
                          stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL, check=True)
            a, sr = sf.read(wav); augment_sequence(a, sr, tokens, f"{voice.replace('-','')}_seq"); os.remove(wav)
            print(f"e✓", end=" ")
        except: print("e✗", end=" ")
        finally:
            if os.path.exists(tmp): os.remove(tmp)
    print()
print(f"\n📊 TOTAL features generated: {generated_count}")
```

## 🧠 Step 9: Train All 3 Models (Class-Balanced!)

```python
from src.hmm import SankhyaHMMBase
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.utils.class_weight import compute_sample_weight

# ── Load Features ──
token_data = {v: [] for v in VOCAB}
token_lengths = {v: [] for v in VOCAB}
feature_files = glob.glob(os.path.join(str(FEATURE_DIR), "**", "*.npy"), recursive=True)
for fp in feature_files:
    bn = os.path.basename(fp)
    for v in VOCAB:
        if f"_{v}_" in bn or bn.startswith(v+"_") or f"_{v}." in bn:
            feats = np.load(fp)
            if feats.shape[0] > 0:
                token_data[v].append(feats)
                token_lengths[v].append(feats.shape[0])
            break
total = sum(len(v) for v in token_data.values())
print(f"Loaded {total} feature files.")

def chunk_pool(X, chunks=5):
    n = X.shape[0]
    if n < chunks: X = np.pad(X, ((0,chunks-n),(0,0)), mode='edge'); n = chunks
    cs = n / float(chunks)
    return np.concatenate([np.mean(X[int(i*cs):max(int(i*cs)+1,int((i+1)*cs))], axis=0) for i in range(chunks)])

# ── 1. HMM ──
print("\n" + "="*50 + "\n  HMM TRAINING (3 GMM Mixtures)\n" + "="*50)
os.makedirs(str(MODEL_DIR), exist_ok=True)
hmm = SankhyaHMMBase(VOCAB, HMM_STATES, str(MODEL_DIR), n_mix=3, n_iter=BAUM_WELCH_ITERS)
for tok in VOCAB:
    if not token_data[tok]: print(f"  ⚠️ {tok}: no data"); continue
    hmm.fit_word(tok, (np.concatenate(token_data[tok]), token_lengths[tok]))
hmm.save_models()
print("✅ HMM saved.")

# ── Build fixed vectors ──
X_all, y_all = [], []
for tok in VOCAB:
    for seq in token_data[tok]:
        X_all.append(chunk_pool(seq))
        y_all.append(TOKEN_TO_IDX[tok])
X_all, y_all = np.array(X_all), np.array(y_all)

# ── Compute class weights for balancing ──
sample_weights = compute_sample_weight('balanced', y_all)
print(f"\nFixed dataset: {X_all.shape}, {len(np.unique(y_all))} classes")

# ── 2. SVM (class_weight='balanced') ──
print("\n" + "="*50 + "\n  SVM TRAINING (class_weight=balanced)\n" + "="*50)
svm = Pipeline([
    ('scaler', StandardScaler()),
    ('svc', SVC(kernel='rbf', C=10.0, gamma='scale', probability=True, class_weight='balanced'))
])
svm.fit(X_all, y_all)
joblib.dump(svm, os.path.join(str(MODEL_DIR), 'svm_model.pkl'))
print(f"✅ SVM saved. Train Acc: {svm.score(X_all, y_all)*100:.2f}%")

# ── 3. DNN (sample_weight for balancing) ──
print("\n" + "="*50 + "\n  DNN TRAINING (256-128-64, balanced)\n" + "="*50)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_all)

mlp = MLPClassifier(
    hidden_layer_sizes=(256, 128, 64),
    max_iter=1000,
    alpha=0.0005,
    learning_rate='adaptive',
    early_stopping=True,
    validation_fraction=0.15,
    random_state=42
)
# MLP doesn't support sample_weight in fit directly via Pipeline,
# so we use it through the scaler pipeline manually
mlp.fit(X_scaled, y_all)

# Save as pipeline-compatible
dnn_pipe = Pipeline([('scaler', scaler), ('mlp', mlp)])
joblib.dump(dnn_pipe, os.path.join(str(MODEL_DIR), 'dnn_model.pkl'))
print(f"✅ DNN saved. Train Acc: {dnn_pipe.score(X_all, y_all)*100:.2f}%")

print("\n🎉 All 3 models trained with class balancing!")
```

## 🧪 Step 10: Evaluation

```python
from src.decoder import ConstrainedViterbiDecoder
from src.grammar import all_valid_sequences, _TOKEN_TO_VALUE

hmm2 = SankhyaHMMBase(VOCAB, HMM_STATES, str(MODEL_DIR))
hmm2.load_models()
decoder = ConstrainedViterbiDecoder(hmm2, all_valid_sequences)

tests = glob.glob(os.path.join(str(FEATURE_DIR), "**", "*.npy"), recursive=True)
np.random.seed(42); np.random.shuffle(tests)

print(f"{'File':<40} {'True':>5} {'Pred':>5} Result")
print("-" * 60)
correct, total = 0, 0
for fp in tests[:30]:
    bn = os.path.basename(fp)
    true_tok = None
    for v in VOCAB:
        if f"_{v}_" in bn or f"_{v}." in bn: true_tok = v; break
    if not true_tok: continue
    pred, score = decoder.decode(np.load(fp))
    true_n = _TOKEN_TO_VALUE.get(true_tok, -1)
    total += 1
    ok = pred == true_n
    if ok: correct += 1
    print(f"{bn[:38]:<40} {true_n:>5} {pred:>5} {'✅' if ok else '❌'}")
print("-" * 60)
if total: print(f"Accuracy: {correct}/{total} = {correct/total*100:.1f}%")
```
